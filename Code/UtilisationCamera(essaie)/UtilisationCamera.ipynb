```python
# Importation de la bibliothèque OpenCV pour la capture vidéo
import cv2
#Importation de matplotlib pour visualisé une image
from matplotlib import pyplot as plt
```

```python
#Connexion et création d'un objet de capture de la webcam en utilisant son identifiant
cap = cv2.VideoCapture(0)
```

```python
#lire une nouvelle image 
ret, frame = cap.read()
```

```python
#permet de voir si l'indice à laquel nous nous somme connecté est celui de la webcam
print (ret)
```

```python
#permet d'affichr les images sous forme de tableau , montrant les valeurs de chaque pixel dans
#chaque canal de couleur(rouge, vert, bleu)
print(frame)
```

```python
#pour visualiser l'image venant de la caméras
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
#permet de fermer la caméra quand on l'exécute
#cap.release()
```

```python
#fonction pour prendre une image 
def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('webcamphoto.jpg', frame)
    cap.release()
```

```python
take_photo() 
```
